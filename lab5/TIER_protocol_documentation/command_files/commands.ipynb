```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chardet
import csv


original_file = '../orginal_data/Datasets/billboard.csv'
```

```python

with open(original_file, 'rb') as f:
    enc = chardet.detect(f.read())

data_from_original_file = pd.read_csv(original_file ,  encoding = enc['encoding'])
pd.DataFrame(data_from_original_file)
```

```python
process_data = '../analysis_data/process_data.csv'
data_from_original_file.to_csv(process_data)
```

```python
df = pd.read_csv(process_data)
pd.DataFrame(df)
```

```python
list_of_column_names = list(df.columns)
# print(list_of_column_names)
for i in list_of_column_names:
    if df[str(i)].isnull().all():
        print("Delated empty columns:", i)
        del df[str(i)]
        list_of_column_names.remove(i)
```

```python
#check are columns are delated
pd.DataFrame(df)
```

```python
df.sort_values(list_of_column_names[8:], axis=0,ascending=True,inplace=True,na_position='first')
pd.DataFrame(df)
```

```python
process_data = '../analysis_data/process_data.csv'
df.to_csv(process_data)
```
