```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Zadanie 3
#Zdefiniuj poni≈ºszƒÖ funkcjƒô i sporzƒÖd≈∫ jej wykres dla argument√≥w z danego przedzia≈Çu:
#f(x) = x**2 +5

def function(x):
    return x**2+5


def plot_functions(start=-1, end=-1, iterations=100, title='Funkcja x**2+5', xlabel='O≈õ x', ylabel='O≈õ y'):
    linspace_x = np.linspace(start, end, iterations)
    y = function(linspace_x)

    plt_title = title + " in range <" + str(start) +":" + str(end) +">"
    plt.plot(linspace_x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(plt_title)
    plt.grid()
    plt.show()

ranges = [[-1, 1], [-6, 6], [0, 5]]

for actual_range in ranges:
    start = actual_range[0]
    end = actual_range[1]
    plot_functions(start, end, 100)
    
```
